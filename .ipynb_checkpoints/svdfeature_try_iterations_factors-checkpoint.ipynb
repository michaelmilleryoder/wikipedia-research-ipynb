```python
import subprocess
import shutil
import os
import pandas as pd
```

```python
# Read in results csv
results = pd.read_csv("/home/michael/software/svdfeature-1.2.2/demo/basicMF/results.csv")
results
```

```python
max_results = []
best = {}
for i, row in results.iterrows():
    if row['#factors'] not in best.keys():
        best[row['#factors']] = (row['#iterations'], row['RMSE'])
    else:
        if row['RMSE'] < best[row['#factors']][1]:
            best[row['#factors']] = (row['#iterations'], row['RMSE'])
            
for f in best:
    max_results.append([f, best[f][0], best[f][1]])

best_results = pd.DataFrame(max_results)
best_results
```

```python
# Change number of iterations in run-wp
n_iters = 200

wpsh = '/home/michael/software/svdfeature-1.2.2/demo/basicMF/run-wp.sh'
shutil.copy(wpsh, wpsh[:-3]+'-old.sh')
with open(wpsh) as f:
    text = f.read().splitlines()

newline8 = "../../svd_feature wp.conf num_round={:d}".format(n_iters)
newline10 = "../../svd_feature_infer wp.conf pred={:d}".format(n_iters)
text[8] = newline8
text[10] = newline10

with open(wpsh, 'w') as f:
    for line in text:
        f.write(line+'\n')
```

```python
# Change number of factors in wp.conf
n_factors = 10

wpconf = '/home/michael/software/svdfeature-1.2.2/demo/basicMF/wp.conf'
shutil.copy(wpconf, wpconf[:-5]+'-old.conf')
with open(wpconf) as f:
    text = f.read().splitlines()

newline17 = "num_factor = {:d}".format(n_factors)
text[17] = newline17

with open(wpconf, 'w') as f:
    for line in text:
        f.write(line+'\n')
```

```python
# Run test, get RMSE from pred.txt output
def run_cmd(cmd):
    output = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    out, err = output.communicate()
    return out, err

os.chdir('/home/michael/software/svdfeature-1.2.2/demo/basicMF/')
out, err = run_cmd('./run-wp.sh')
print(out)
```
